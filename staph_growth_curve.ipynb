```python
import matplotlib.pyplot as plt
```

```python
# Data
time = [0, 1, 2, 3, 4, 6, 8, 10, 14, 18, 24, 28, 32, 36]
log_cfu = [4.0, 4.02, 4.1, 4.8, 5.8, 7.5, 8.6, 9.0, 9.1, 9.0, 8.8, 7.8, 6.5, 5.0]
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

# Phase background colors
ax.axvspan(0, 2, alpha=0.2, color='red', label='Lag Phase')
ax.axvspan(2, 10, alpha=0.2, color='green', label='Exponential Phase')
ax.axvspan(10, 24, alpha=0.2, color='yellow', label='Stationary Phase')
ax.axvspan(24, 36, alpha=0.2, color='gray', label='Death Phase')

# Plot the growth curve
ax.plot(time, log_cfu, 'ko-', linewidth=2, markersize=6)

ax.set_xlabel('Incubation Time (Hours)', fontsize=12)
ax.set_ylabel('Log CFU/mL', fontsize=12)
ax.set_title('Growth phases of S. aureus', fontsize=14, fontweight='bold')
ax.set_xlim(0, 36)
ax.set_ylim(3, 10)
ax.legend(loc='lower right')
ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
